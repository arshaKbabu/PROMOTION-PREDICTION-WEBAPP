```python
import pandas as pd
import  numpy as np

from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from word2number import w2n

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

```

```python
Data = Data.drop('gender',axis=1)
Data
```

```python
dept_map={'Finance':0,'HR':1,'IT':2,'Marketing':3}
Data['department']=Data['department'].replace(dept_map)
Data
```

```python
edu_map={'Bachelors':0,'High School':1,'Masters':2,'PhD':3}
Data['education']=Data['education'].map(edu_map)
Data
```

```python
x=Data.drop(['promotion'],axis=1)
y=Data['promotion']
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

# TRAINING MODELS ONE BY ONE

```python
from sklearn.metrics import confusion_matrix, classification_report

# Dictionary of all models
models = {
    "Logistic Regression": LogisticRegression(),
    "Gaussian Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN (k=15)": KNeighborsClassifier(n_neighbors=15),
    "SVM": SVC(random_state=42),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

results=[]
# Loop through all models
for name, model in models.items():
    print(f"\n==============================")
    print(f"Model: {name}")
    print("==============================")

    # Train
    model.fit(X_train, y_train)

    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Scores
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)

    print(f"Train Accuracy: {train_score:.4f}")
    print(f"Test Accuracy: {test_score:.4f}")

    # Confusion Matrix & Classification Report
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_test_pred))

    print("\nTrain Classification Report:")
    print(classification_report(y_train, y_train_pred))

    print("\nTest Classification Report:")
    print(classification_report(y_test, y_test_pred))


    # Append to results
    results.append([
        name,
        train_score, test_score,
    ])


# Create DataFrame
df_results = pd.DataFrame(results, columns=[
    "Model",
    "Train Accuracy", "Test Accuracy"
])


```

# HYPER PARAMETER TUNING

```python
model_params = {
    
    "Logistic Regression": {
        "model": LogisticRegression(max_iter=500),
        "params": {
            "C": [0.1, 1, 10],                     # default = 1.0
            "penalty": ["l2"],                    # default = 'l2'
            "solver": ["lbfgs", "liblinear"]      # default = 'lbfgs'
            # max_iter default = 100
        }
    },

    "Gaussian Naive Bayes": {
        "model": GaussianNB(),
        "params": {
            "var_smoothing": [1e-09, 1e-08, 1e-07]  # default = 1e-9
        }
    },

    "Decision Tree": {
        "model": DecisionTreeClassifier(random_state=42),
        "params": {
            "criterion": ["gini", "entropy"],       # default = 'gini'
            "max_depth": [3, 5, 10],                # default = None
            "min_samples_split": [2, 5, 10]         # default = 2
        }
    },

    "KNN (k=15)": {
        "model": KNeighborsClassifier(),
        "params": {
            "n_neighbors": [5, 10, 15],             # default = 5
            "weights": ["uniform", "distance"],     # default = 'uniform'
            "p": [1, 2]                              # default = 2 (Euclidean)
        }
    },

    "SVM": {
        "model": SVC(),
        "params": {
            "C": [1, 10, 20],                       # default = 1.0
            "kernel": ["rbf", "linear"],            # default = 'rbf'
            "gamma": ["auto", "scale"]              # default = 'scale'
        }
    },

    "Random Forest": {
        "model": RandomForestClassifier(),
        "params": {
            "n_estimators": [50, 100, 200],         # default = 100
            "criterion": ["gini", "entropy"],       # default = 'gini'
            "max_depth": [None, 5, 10]              # default = None
        }
    },

    "AdaBoost": {
        "model": AdaBoostClassifier(),
        "params": {
            "n_estimators": [50, 100, 200],         # default = 50
            "learning_rate": [0.5, 1.0, 2.0],       # default = 1.0
            "algorithm": ["SAMME", "SAMME.R"]       # default = 'SAMME.R'
        }
    },

    "Gradient Boosting": {
        "model": GradientBoostingClassifier(),
        "params": {
            "n_estimators": [50, 100],              # default = 100
            "learning_rate": [0.05, 0.1, 0.2],      # default = 0.1
            "max_depth": [2, 3, 5]                  # default = 3
        }
    }
}
```

```python
# Run GridSearchCV for all models
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp["model"], mp["params"], cv=5, return_train_score=True)
    clf.fit(X_train, y_train)

    scores.append({
        "model": model_name,
        "best_score": clf.best_score_,
        "best_params": clf.best_params_
    })

# Convert to DataFrame
df = pd.DataFrame(scores)

```
